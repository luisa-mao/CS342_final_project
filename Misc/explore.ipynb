```python
import pandas as pd
```

```python
big_df = pd.DataFrame()

i = 0
while True:
    # pad i to 5 digits
    i_str = str(i).zfill(5)
    print(i_str)
    df = None
    try:
        # read the parquet file
        df = pd.read_parquet(f'/home/amcalinden/cc3m_train/{i_str}.parquet')
        df = df[df['status'] == 'success']
        df = df[['caption', 'key']]
        big_df = pd.concat([big_df, df], ignore_index=True)
    except FileNotFoundError:
        print("Ended at ", i_str)
        break
    i += 1

print(big_df.shape)
print(big_df.head())
```

```python
print(big_df.shape)
print(big_df.head())
# write the dataframe to a csv file
big_df.to_csv('cc3m_train.csv', index=False)
```
