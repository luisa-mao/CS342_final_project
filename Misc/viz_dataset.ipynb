```python
import os
import sys
parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(parent_dir)
wnids_path = os.path.join(parent_dir, "tiny-imagenet-200/wnids.txt")
words_path = os.path.join(parent_dir, "tiny-imagenet-200/words.txt")

# Load wnids.txt (200 valid classes)
with open(wnids_path, 'r') as f:
    valid_wnids = set(f.read().splitlines())

wnid_to_word = {k:[] for k in valid_wnids}

# Load words.txt and filter only valid wnids
filtered_words = []
with open(words_path, 'r') as f:
    for line in f:
        wnid, description = line.split('\t')
        if wnid in valid_wnids:
            words = description.split(',')
            filtered_words.append(words)
            wnid_to_word[wnid] += words

# filtered_words now contains only valid wnids from wnids.txt
print(wnid_to_word['n02909870'])
```
